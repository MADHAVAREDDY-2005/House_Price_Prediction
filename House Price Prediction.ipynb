```python
colss=['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']
outliers={}
```

```python
# Finding Outliers using IQR method
for col in colss:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_rows = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers[col] = outlier_rows.shape[0]
    print(f"{col}: {outlier_rows.shape[0]} outliers found")
```

```python
# Outliers using Z-scores
import numpy as np
from scipy.stats import zscore
z_scores = np.abs(df[colss].apply(zscore))
outliers_z = (z_scores > 3)
for col in colss:
    print(f"{col}: {outliers_z[col].sum()} outliers found")
```

```python
# Outliers using Mod_Z-Score
def modified_z_score(series):
    median = np.median(series)
    mad = np.median(np.abs(series - median))  # Median Absolute Deviation
    modified_z = 0.6745 * (series - median) / mad
    return np.abs(modified_z)

modified_z_scores = df[colss].apply(modified_z_score)
outliers_modified = (modified_z_scores > 3.5)
for col in colss:
    print(f"{col}: {outliers_modified[col].sum()} outliers found")
```
